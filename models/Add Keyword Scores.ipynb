```python
file1 = open('sorted_keyword_distribution.csv', 'r')
Lines = file1.readlines()

corpus_old = []
keyword_score_map = {}
for line in Lines:
    try:
        arr = line.strip().split(",")
        key_arr = arr[:-1]
        key = ""
        for keyword in key_arr:
            key += keyword + ","
        key = key[:-1]
        keyword_score_map[key] = int(arr[-1])
    except:
        continue
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="",
  database="extracted_keyword_validation"
)
mycursor = mydb.cursor()
```

```python
with open("keyword_score_map.pkl", 'wb') as output:  # Overwrites any existing file.
    pickle.dump(keyword_score_map, output, pickle.HIGHEST_PROTOCOL)
```
